```python
attack_mean=data["Attack"].mean()

def set_attack(val):
    if val< attack_mean:
        return "Attack Low"
    elif val==attack_mean:
        return "Attack neutral"
    else:
        return "Attack High"
```

```python
data["Attack_high_low"]=data["Attack"].apply(set_attack)
data
```

```python
speed_mean=data["Speed"].mean()

def set_speed(val):
    if val< speed_mean:
        return "Speed Low"
    elif val==speed_mean:
        return "Speed neutral"
    else:
        return "Speed High"
```

```python
data["Speed_high_low"]=data["Speed"].apply(set_speed)
data
```

```python
hp_mean=data["HP"].mean()

def set_hp(val):
    if val< hp_mean:
        return "HP Low"
    elif val==hp_mean:
        return "HP neutral"
    else:
        return "HP High"
```

```python
data["HP_high_low"]=data["HP"].apply(set_hp)
data
```
